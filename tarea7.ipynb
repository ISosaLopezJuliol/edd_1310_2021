```
class NodoDoble:
  def __init__ (self , value, siguiente= None , anterior=None):
    self.data= value
    self.next= siguiente
    self.prev= anterior

class DoubleLinkedList:
  def __init__( self ):
    self.__head = NodoDoble( None , None , None )
    self.__tail = NodoDoble( None , None , None )
    self.__head.next = self.__tail 
    self.__tail.prev = self.__head 
    self.__size = 0
    print(f"La cabeza {self.__head.data}")
    print(f"La cabeza {self.__tail.data}")


  def is_empty( self ):
        return self.__head.data == None
  
  def append( self , value ):
        curr_node = NodoDoble( value )
        if self.is_empty() == True:
            self.__head = curr_node
            self.__tail= curr_node
        else:
             curr_node.next = self.__tail
             self.__tail.prev= curr_node
             self.__tail = curr_node 
            
  
  def transversal( self ):
        curr_node = self.__head
        while curr_node != None:
            print(f" { curr_node.data } --> " , end="")
            curr_node = curr_node.prev
        print(" ")
    
  def reverse_tranversal(self):
    curr_node = self.__tail
    while curr_node !=None:
      print(f"{ curr_node.data} --> " , end="")
      curr_node = curr_node.next
    print(" ")


  def tail ( self ):  
        curr_node = self.__head
        while curr_node.prev != None: #recorre todo hasta encontrar un None y regresa uno 
            curr_node = curr_node.prev
        return curr_node

  def head ( self ):  
        curr_node = self.__tail
        while curr_node.next != None: #recorre todo hasta encontrar un None y regresa uno 
            curr_node = curr_node.next
        return curr_node
  
  def remove_from_tail( self , value ):  
        curr_node = self.__tail
        if self.__tail.data == value:
            self.__tail = self.__tail.next
            self.remove_from_head(curr_node.data)
        else:
            anterior = None 
            while curr_node.data != value and curr_node.next != None:
                anterior = curr_node
                curr_node= curr_node.next
                if curr_node.data == value:
                 anterior.next= curr_node.next
                 self.remove_from_head(curr_node.data)

            if self.__head.data == value:
                self.__tail = self.tail()

  def remove_from_head( self , value ): 
        curr_node = self.__head
        if self.__head.data == value:
            self.__head = self.__head.prev
            self.remove_from_tail(curr_node.data)
        else:
            anterior = None
            while curr_node.data != value and curr_node.prev != None:
                anterior = curr_node
                curr_node= curr_node.prev
                if curr_node.data == value:
                 anterior.prev= curr_node.prev 
                 self.remove_from_tail(curr_node.data)    

  def size(self): 
      curr_node = self.__head
      while curr_node !=None:
           curr_node= curr_node.prev
           self.__size = self.__size + 1  
      print(f"La lista tine {self.__size} nodos")  
    
  def get (self, posicion= None): #regresa el ultimo 
        contador=0
        dato= None
        if posicion == None:
            dato = print("ingresa la pocisión de un nodo")
        else:
            curr_node = self.__head
            antes = None
            while contador != posicion and curr_node.prev != None:
                contador = contador + 1
                antes = curr_node
                curr_node = curr_node.prev
            if contador == posicion:
                dato = print(f"El valor del nodo pedido es de {curr_node.data}")
            else:
                dato = print("El nodo pedido no existe ingrese otro")

  def find_form_tail(self, value= None): 
        contador=0
        dato= None
        if value == None:
            dato = None
        else:
            curr_node = self.__tail
            anterior = None 
            while curr_node.data != value and curr_node.next != None:
                anterior = curr_node
                curr_node= curr_node.next
                contador= contador + 1
            if curr_node.data == value:
                 dato = dato= print(f"El valor esta en la posición: {contador}") 
            else: 
                 dato = print("El valor no existe en la lista")
        return dato
 
  def find_form_head(self, value= None): 
        contador=0
        dato= None
        if value == None:
            dato = print("no ingresate un valor")
        else:
            curr_node = self.__head
            anterior = None 
            while curr_node.data != value and curr_node.prev != None:
                anterior = curr_node
                curr_node= curr_node.prev
                contador= contador + 1
            if curr_node.data == value:
                 dato = dato= print(f"El valor esta en la posición: {contador}") 
            else: 
                 dato = print("El valor no existe en la lista")
        return dato

l = DoubleLinkedList()      
l.append(2)
l.append(4)
l.append(8)
l.append(16)
l.append(32)
l.transversal()
l.reverse_tranversal()
print(" se removio un numero")
l.remove_from_head(2)
l.transversal()
l.reverse_tranversal()
print("")
l.remove_from_tail(32)
l.transversal()
l.reverse_tranversal()
l.size()
print("")
l.get(4)
l.find_form_tail(8)
l.find_form_head(8)
```
